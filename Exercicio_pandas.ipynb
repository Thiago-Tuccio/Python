Visualização dos produtos

```python
#Saber quais são os possíveis produtos
Produtos = dados['Produto'].unique()

#percorrer a lista de possiveis produtos
for i in range(0,len(Produtos)):
    print(Produtos[i])
```

Função auxiliar que calcula a média ponderada

```python
def med_pond_b(lista, qtd):
    if len(lista) != 0: #Para corrigir problema de o operador não operar o produto, pois ele manda uma lista vazia (talvez aliminar listas vazias seja uma opção)
        soma  = 0
        soma_qtd = 0
        for n in range(len(lista)): 
            soma = soma + lista[n]*qtd[n]
            soma_qtd += qtd[n] 
            media = soma/soma_qtd
        print("A média de preço de compra para", Produtos[i] ,"é: ", media)


def med_pond_s(lista, qtd):
    if len(lista) != 0: #Para corrigir problema de o operador não operar o produto, pois ele manda uma lista vazia (talvez aliminar listas vazias seja uma opção)
        soma  = 0
        soma_qtd = 0
        for n in range(len(lista)): 
            soma = soma + lista[n]*qtd[n]
            soma_qtd += qtd[n]
            media = soma/soma_qtd
        print("A média de preço de venda para", Produtos[i] ,"é: ", media)
```

Preços médios ponderados para a Casa A

```python
#percorrer a lista de possiveis produtos
for i in range(0,len(Produtos)):
    CASA_A_PROD = CASA_A.loc[CASA_A['Produto'] == Produtos[i]] #Cria um DF para cada produto que a Casa A operou
    
    CASA_A_PROD_FUTURE = CASA_A_PROD.loc[CASA_A['Product Type'] == 'Future'] #Pega só o futuro para cada produto que a Casa A operou e cria um DF
    
    #Dividisão em compra (B) e venda (S)
    #Compra (B)
    CASA_A_PROD_FUTURE_BUY = CASA_A_PROD_FUTURE.loc[CASA_A['Compra/Venda'] == 'B']
    #display(CASA_A_PROD_FUTURE_BUY)
    lista_b = CASA_A_PROD_FUTURE_BUY['Price'].tolist()
    qtd_b = CASA_A_PROD_FUTURE_BUY['Qty'].tolist()
    med_pond_b(lista_b, qtd_b)


    #Venda (S)
    CASA_A_PROD_FUTURE_SELL = CASA_A_PROD_FUTURE.loc[CASA_A['Compra/Venda'] == 'S']
    #display(CASA_A_PROD_FUTURE_SELL)
    lista_s = CASA_A_PROD_FUTURE_SELL['Price'].tolist()
    qtd_s = CASA_A_PROD_FUTURE_SELL['Qty'].tolist()
    med_pond_s(lista_s, qtd_s)
```

```python
#percorrer a lista de possiveis produtos
for i in range(0,len(Produtos)):
    CASA_B_PROD = CASA_B.loc[CASA_B['Produto'] == Produtos[i]] #Cria um DF para cada produto que a Casa A operou
    
    CASA_B_PROD_FUTURE = CASA_B_PROD.loc[CASA_B['Product Type'] == 'Future'] #Pega só o futuro para cada produto que a Casa A operou e cria um DF
    
    #Dividisão em compra (B) e venda (S)
    #Compra (B)
    CASA_B_PROD_FUTURE_BUY = CASA_B_PROD_FUTURE.loc[CASA_B['Compra/Venda'] == 'B']
    #display(CASA_A_PROD_FUTURE_BUY)
    lista_b = CASA_B_PROD_FUTURE_BUY['Price'].tolist()
    qtd_b = CASA_B_PROD_FUTURE_BUY['Qty'].tolist()
    med_pond_b(lista_b, qtd_b)


    #Venda (S)
    CASA_B_PROD_FUTURE_SELL = CASA_B_PROD_FUTURE.loc[CASA_B['Compra/Venda'] == 'S']
    #display(CASA_A_PROD_FUTURE_SELL)
    lista_s = CASA_B_PROD_FUTURE_SELL['Price'].tolist()
    qtd_s = CASA_B_PROD_FUTURE_SELL['Qty'].tolist()
    med_pond_s(lista_s, qtd_s)
```
