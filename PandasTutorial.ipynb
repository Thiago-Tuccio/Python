```python
notas = pd.DataFrame({
    "Nome": ["Ana","Pedro", "João"],
    "Nota 1": [9,7,10],
    "Nota 2": [6,7,9],
    "Nota 3": [8,4,10],
    "Nota 4": [5,8,8]})
print(notas)

#agora vou calcular a média, fazendo assim ele adiciona uma nota coluna ao DataFrame
notas["Médias"] = (notas["Nota 1"] + notas["Nota 2"] + notas["Nota 3"] + notas["Nota 4"])/4
print("\n", notas)

#agora criando uma coluna com um valor padrão, adiciona o mesmo valor a todas as linhas
notas["Faltas"] = 3
print("\n", notas)
```

Tratamento de Dados

```python
Dados_DF = pd.read_excel("Tratamento_Dados.xlsx")


#O método .fillna vai operar sobre os valores vazios daquela coluna selecionada
#Dados_DF["Total Vendas"].mean() - calcula a média dos valores da coluna mean() exclui automaticamente os valores NaN (não é um número) ao calcular a média.
Dados_DF["Total Vendas"] = Dados_DF["Total Vendas"].fillna(Dados_DF["Total Vendas"].mean())

#Se eu quiser colocar um valor fixo nos lugares vazios uso o .fillna assim:
Dados_DF["Total Vendas"] = Dados_DF["Total Vendas"].fillna(500)

# o método .ffill preenche os valores com o ultimo registro válido
Dados_DF["Total Vendas"] = Dados_DF["Total Vendas"].ffill()

Dados_DF
```

Merge INNER JOIN

```python
Loja1_DF = pd.read_excel("Vendas_INNER_JOIN_Loja1.xlsx")
Loja2_DF = pd.read_excel("Vendas_INNER_JOIN_Loja2.xlsx")

#on - em qual coluna fazer o merge
#how - como fazer o merge
#"inner" - faz o merge entre as tabelas
#suffies=("_Loja1", "_Loja2") - vai colocar as legendas no cabeçalho da tabela, se nao colocar fixa X e Y
VendedoresAmbasLojas_DF = pd.merge(Loja1_DF, Loja2_DF, on="Vendedor", how="inner", suffixes=("_Loja1", "_Loja2"))

#Podemos Filtrar os dados do Merge da seguinte maneira, especificando qual coluna queremos de qual DF queremos Loja1_DF, Loja2_DF[["Vendedor", "Total Vendas"]]
VendedoresAmbasLojas_DF = pd.merge(Loja1_DF, Loja2_DF[["Vendedor", "Total Vendas"]], on="Vendedor", how="inner", suffixes=("_Loja1", "_Loja2"))

VendedoresAmbasLojas_DF
```

Merge Left Join - Conferir se uma pessoa aparece em uma base de dados (ou alguma coisa)

```python
Vendas_DF = pd.read_excel("Vendas_LEFT_JOIN.xlsx")
Vendedores_DF = pd.read_excel("Vendedores_LEFT_JOIN.xlsx")

#on="Id Vendedor" - coluna que vai procurar
#how="left" - para por na esquerda
#O parâmetro how="left" em uma operação de mesclagem (merge) indica que você está interessado em manter todas as linhas do dataframe à esquerda (no dataframe Vendas_DF) e mesclar com as linhas correspondentes no dataframe à direita (no dataframe Vendedores_DF).
#Se houver linhas em Vendas_DF que não tenham uma correspondência na coluna Id Vendedor em Vendedores_DF, essas linhas permanecerão no dataframe resultante (VerificandoVendas_DF), com os valores das colunas do dataframe Vendedores_DF preenchidos com NaN (Not a Number).
#Essencialmente, isso é útil quando você deseja manter todos os registros do dataframe da esquerda, mesmo que não haja correspondências no dataframe da direita.
VerificandoVendas_DF = pd.merge(Vendas_DF, Vendedores_DF, on=["Id Vendedor"], how="left", suffixes=("Vendas","Checagem"))


#.dropna() deleta todas as linhas com pelo menos um valor vazio
LimpandoLinhasNaN = VerificandoVendas_DF.dropna()

#Deletando a coluna "VendedorChecagem"
del LimpandoLinhasNaN["VendedorChecagem"]
VerificandoVendas_DF
LimpandoLinhasNaN
```

Groupby - agrupa o DF por alguma variável de interesse

```python
Vendas_DF = pd.read_excel("Groupby.xlsx")

#.groupby - agrupando pela coluna de vendedores
#mean - média excluindo NaN
#dropna=False - para não descartar valores em branco
#.drop(columns=["Produto"]) para excluir a coluna produto \ atualização do pandas não deixa mais fazer sem isso
MediaVendedor = Vendas_DF.drop(columns=["Produto"]).groupby(["Vendedor"], dropna=False).mean()

#.groupby(["Vendedor", "Produto"] - assim ele agrupa duas colunas
MediaProdutoVendedor = Vendas_DF.groupby(["Produto", "Vendedor"], dropna=False).mean()

display(MediaProdutoVendedor)
```

Pivot Table

```python
Vendas_DF = pd.read_excel("Vendas_Lanchonete_Pivot_Table.xlsx")

#index="Data Venda" - se tornará a linha do novo DF
#columns="Cliente" - colunas do novo DF
# values=['Preço com Desconto','Preço Total'] - valores que quero observar da relação entre index="Data Venda" e columns="Cliente", aceita mais de 1
#.pivot_table - reorganizar os dados de um DataFrame e aceita linhas duplicadas / valores repetidos na coluna
#aggfunc="sum" - Soma os valores , sem ela os valores vem como média
pivotTableExemplo1 = Vendas_DF.pivot_table(index="Data Venda", columns="Cliente", values = ['Preço com Desconto','Preço Total'], aggfunc="sum")

#index="Data Venda" - se tornará a linha do novo DF
#columns=["Cliente", 'Produto'] - colunas do novo DF, para cada cliente os produtos
# values=['Preço Total'] - valores que quero observar da relação entre index="Data Venda" e columns="Cliente", aceita mais de 1
#.pivot_table - reorganizar os dados de um DataFrame e aceita linhas duplicadas / valores repetidos na coluna
#aggfunc="sum" - Soma os valores , sem ela os valores vem como média
pivotTableExemplo2 = Vendas_DF.pivot_table(index="Data Venda", columns=["Cliente", 'Produto'], values = ['Preço Total'], aggfunc="sum")

#Também posso fazer uma combinação dos dois anteriores
pivotTableExemplo3 = Vendas_DF.pivot_table(index="Data Venda", columns=["Cliente", 'Produto'], values = ['Preço com Desconto','Preço Total'], aggfunc="sum")

#Para substiruir o NaN por 0
#só no exemplo 3
pivotTableExemplo3["Preço Total"] = pivotTableExemplo3["Preço Total"].fillna(0)
pivotTableExemplo3["Preço com Desconto"] = pivotTableExemplo3["Preço com Desconto"].fillna(0)

display(pivotTableExemplo3)
```

Quabrando um arquivo em vários e gerando CSV

```python
baseDados_DF = pd.read_excel("Vendas_Jan.xlsx")

#subset - Coluna que queremos remover
#keep - Instrução do qual valor vai permanecer (first / primeiro)
removendoDuplicidades = baseDados_DF.drop_duplicates(subset="Vendedor", keep="first")

for linha in removendoDuplicidades["Vendedor"]:
    
    #loc - Localizar
    vendas_Funcionario = baseDados_DF.loc[baseDados_DF["Vendedor"] == linha]
    
    #Salvando o arquivo em formato excel csv
    vendas_Funcionario.to_csv("Relatório Vendas " + linha + ".csv")
```

Quebrando um arquivo em vários e gerando XLSX

```python
baseDados_DF = pd.read_excel("Vendas_Jan.xlsx")

# Remover duplicidades com base na coluna "Vendedor"
removendoDuplicidades = baseDados_DF.drop_duplicates(subset="Vendedor", keep="first")

# Iterar sobre cada vendedor na base de dados sem duplicatas
for linha in removendoDuplicidades["Vendedor"]:
    
    # Filtrar as vendas para o vendedor atual
    vendas_Funcionario = baseDados_DF.loc[baseDados_DF["Vendedor"] == linha]
    
    # Criar um objeto para escrever no arquivo Excel
    arquivoExcel = pd.ExcelWriter("Relatório Vendas " + linha + ".xlsx", engine="xlsxwriter")
    
    # Transformar os dados do vendedor em um DataFrame
    dataFrame = pd.DataFrame(vendas_Funcionario)
    
    # Preparar o arquivo Excel
    #index=False -  não coloca o índice do DF no excel
    #sheet_name="Dados" - nome da página da planilha do excel
    dataFrame.to_excel(arquivoExcel, sheet_name="Dados", index=False)
    
    # Salvar o arquivo Excel
    arquivoExcel.close()
```
