```python
import pandas as pd
import matplotlib.pyplot as plt

#Bibliotecas de Finanças
import yfinance as yf
import fundamentus
from tradingcomdados import alternative_data as ad
from tradingcomdados import funds_data as fd

#Bibliotecas para dashboard
import vizro.plotly.express as px
from vizro import Vizro
import vizro.models as vm
```

# Importação de Dados

```python
df = fundamentus.get_resultado()
df['TICKER'] = df.index.str[:4] #Pegando o texto do Ticker do ativo
df['FINAL'] = df.index.str[4:6] #Pegando o número final do Ticker
df
```

```python
setores = ad.get_sectors()
setores.rename(columns={'CÓDIGO':'TICKER', 'NOME NO PREGÃO':'EMPRESA'}, inplace=True)
setores
```

```python
resultados = pd.merge(df, setores, how ='inner', on ='TICKER')
resultados
```

# Criando Dashboard

```python
page = vm.Page(
    title='Dashboard',
    components = [
        vm.Graph(id='Dividend Yield', figure=px.bar(resultados, x='TICKER', y='dy', color='SETOR ECONÔMICO')),
        vm.Graph(id='Preço/Valor Patrimonial', figure=px.scatter(resultados, x='TICKER', y='pvp', color='SETOR ECONÔMICO')),
        vm.Graph(id='Return on Equity', figure=px.scatter(resultados, x='TICKER', y='roe', color='SETOR ECONÔMICO'))
    ],
    controls = [
        vm.Filter(column='SEGMENTO'),
        vm.Filter(column='pvp', selector=vm.RangeSlider()),
        vm.Filter(column='roe', selector=vm.RangeSlider()),
        vm.Filter(column='dy', selector=vm.RangeSlider())    
    ]
)

dashboard = vm.Dashboard(pages=[page])

Vizro().build(dashboard).run()
```
