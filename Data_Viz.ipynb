# Bibliotecas

```python
import yfinance as yf
import pygwalker as pyg
import wbgapi as wb
import pandas as pd
import numpy as np
```

# Dados

```python
df = yf.download('PETR4.SA', start='2010-01-01')
```

```python
#Retorno em %
df['Retorno'] = df['Adj Close'].pct_change()

#Volatilidade dentro de uma janela de 10 dias
df["Vol_10_dias"] = df['Retorno'].rolling(10).std() #.std é o desvio padrão que está sendo usado como medida de risco

#Media móvel dentro de 2o dias
df['MediaMovel_20_dias'] = df['Adj Close'].rolling(20).mean()

#Ano e mês
df['Ano'] = df.index.year
df['Mês'] = df.index.month
df['Date'] = df.index
df
```

# PYGWALKER - Dashboard

```python
walker = pyg.walk(df)
```

WBGAPI

```python
paises = ['ARG', 'BRA', 'CHL', 'COL', 'PER', 'URY', 'MEX']

desemprego = wb.data.DataFrame('SL.UEM.TOTL.ZS', economy=paises, time=range(2010,2022))
desemprego = desemprego.melt(var_name='Ano',value_name='Desemprego')
desemprego['País'] = np.resize(paises, len(desemprego))
desemprego['Ano'] = pd.to_numeric(desemprego['Ano'].str.slice(2)) #.str.slice(2) tira os dois primeiros elementos da string
desemprego['Desemprego'] = pd.to_numeric(desemprego['Desemprego'])

pyg.walk(desemprego)
desemprego
```
